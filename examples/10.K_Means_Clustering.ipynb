# K-means clustering (non-OOP implementation)
In this notebook, we run k-means clustering on the Iris CSV data to perform unsupervised learning.

Import libraries and read the data.

```python
import pandas
import sklearn
import sklearn.cluster
data = pandas.read_csv("data/Iris.csv").iloc[:,1:-1]
data
```

Run k-means and print the inertia (total loss).

```python
from matplotlib import pyplot

centroid, labels, inertia = sklearn.cluster.k_means(data, 3)
inertia
```

Plot all combinations of variables with clustering colour-coded.

```python
import itertools

figure = pyplot.figure(figsize=(30,30))
for plot, columns in enumerate(itertools.combinations(range(4), 3)):
    axes = figure.add_subplot(1, 4, plot + 1, projection="3d")
    for f, i in zip((axes.set_xlabel, axes.set_ylabel, axes.set_zlabel), columns):
        f(data.columns[i])
    axes.scatter(*(data.iloc[:, i] for i in columns), c=labels)
    axes.scatter(*(centroid.T[i] for i in columns),c="red",s=500)

figure.tight_layout()
```
